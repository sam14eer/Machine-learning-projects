## Importing the required laibraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
# importing data and creating the dataframe
credit_card_data=pd.read_csv('creditcard.csv')
```

```python
# getting the basic information of data like columns,datatypes,null values and total values
credit_card_data.info()
```

```python
# checking for null values in the dataset
credit_card_data.isna().sum()
```

```python
# checking the legit and fraudulent transactions
credit_card_data['Class'].value_counts()
```

## as we can see the data is highly imbalanced

```python
#distributing the data in legit and fraude based on class values
legit=credit_card_data[credit_card_data.Class==0]
fraude=credit_card_data[credit_card_data.Class==1]
```

```python
legit.shape
```

```python
fraude.shape
```

```python
legit['Amount'].describe()
```

```python
fraude['Amount'].describe()
```

```python
# comparing the values of the both transactions
credit_card_data.groupby('Class').mean()
```

```python
# building a dataset in same praportion for legit and fraude transaction to deal with unbalanced data by using under sampeling
legit_sample=legit.sample(n=492)
```

```python
# concatinating the megit dataframe with fraude data frame to make a new data which is balance
new_data=pd.concat([legit_sample,fraude],axis=0)
```

```python
new_data.head()
```

```python
new_data['Class'].value_counts()
```

```python
new_data.groupby('Class').mean()
```

```python
# splitting the data into featires and target
x=new_data.drop(columns=('Class'),axis=1)
y=new_data['Class']
```

```python
x
```

```python
y
```

```python
# splitting the data set into training data and test data
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

## Building the classification model

```python
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
# training data accuracy
training_prediction=model.predict(X_train)
```

```python
training_accuracy=accuracy_score(training_prediction,Y_train)
print('traning data accuracy of the model is =',training_accuracy)
```

```python
# test data accuracy
test_prediction=model.predict(X_test)
```

```python
test_accuracy=accuracy_score(test_prediction,Y_test)
print('test data accuracy of the model is =',test_accuracy)
```
